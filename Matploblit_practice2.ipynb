```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create scatterplot using o
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatterplot using scatter()
plt.scatter(x_axis, y_axis)
```

```python
# Create scatterplot using scatter() 
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Dates")
# Set the x limit between 0 and 45.
plt.xlim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a Bubble Chart using MATLAB
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Same as above just using list 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create scatteplot using Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart using Object_Oriented Interface
fig, ax = plt.subplots()
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*5)
ax.scatter(y_axis, x_axis, s=y_axis_larger, color="skyblue", edgecolor="black", linewidth=2, label= "Boston")
fig.gca().invert_yaxis()
ax.set_xlim(0, 45)
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Dates')
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create same pie chart using >>>> Object-Oriented Interface
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```
